```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

plt.style.use('ggplot')
rcParams['figure.figsize'] = 13, 7
rcParams['figure.figsize'] = 13, 7
rcParams["xtick.labelsize"] = 16
rcParams["ytick.labelsize"] = 16
rcParams["axes.labelsize"] = "x-large"
%matplotlib inline
```

```python
def lookup(s):
    """
    speed up date-parse
    """
    dates = {date:pd.to_datetime(date, errors='coerce') for date in s.unique()}
    return s.map(dates)

def barplot(df, attribute, top=5):
    fig, ax = plt.subplots()
    df_plt = df.groupby([attribute]).size().sort_values(ascending=False
                                                       ).head(top).sort_values(
                                                        ascending=True)
    df_plt.plot(kind="barh", ax=ax)
    plt.xlabel("No. of Cases")
    
```

#### Timeseries perspective 

```python
def make_time_variables(df):
    columns = df.select_dtypes(include=['datetime64']).columns.values
    for col in columns:
        new_col_w = col+"_w"
        new_col_m = col+"_m"
        df[new_col_w] = clean[col].dt.weekday
        df[new_col_m] = clean[col].dt.month
    return df
```

```python
time = clean.copy()
time = make_time_variables(time)
```

```python
time.columns.values
```

```python

f, ax = plt.subplots(2, sharex=True)


tf = pd.DataFrame(time.groupby(["Incident Date_m", "Severity"]
                                     ).size(
                                        ).unstack().fillna(0))[['F  ', 'M  ', 'O  ']]
tf_norm_s = ((tf - tf.mean()) / (tf.max()) )
tf_norm_s.plot(ax=ax[0])

ax[0].set_title('Incident month trend (up) v.s. Dispo month trend (down)')
ax[0].legend(loc='center left', bbox_to_anchor=(1, 0.5))
tf = pd.DataFrame(time.groupby(["Dispo Date_m", "Severity"]
                                     ).size(
                                        ).unstack().fillna(0))[['F  ', 'M  ', 'O  ']]
tf_norm = ((tf - tf.mean()) / (tf.max()) )
tf_norm.plot(ax=ax[1])

ax[1].set_xticks([i for i in range(1,13)])
ax[1].set_xlabel("Month")
ax[0].yaxis.set_label_coords(-0.15, -0.2)
ax[0].set_ylabel("Normalized No. of Cases",)
ax[1].legend(loc='center left', bbox_to_anchor=(1, 0.5))
#plt.title("Monthly Trend for different Severity level [based on incident date]")
```

#### Data Engineering for spatial visualization

```python
## output the data for map
spatial_visaul = clean[["Incident Zip","Severity", "Class"]].dropna()
spatial_visaul["new_set"] = spatial_visaul["Severity"].str.strip() + "_" + \
spatial_visaul["Class"]
spatial_visaul.groupby(["Incident Zip", "new_set"]).size(
).unstack().fillna(0).to_csv("spatial_visual.csv")
```

```python
output2 = pd.DataFrame(spatial_visaul.groupby(["Incident Zip", "new_set"]).size())
```

```python
output2 = output2.reset_index()
```

```python
idx = output2.groupby(['Incident Zip'])[0].transform(max) == output2[0]
```

```python
outputfile = output2[output2.columns.values[:2]][idx]
```

```python
def replace_str(x):
    return "https://www.unitedstateszipcodes.org/{}/#stats".format(str(x))

outputfile["description"] = outputfile["Incident Zip"].apply(replace_str)
```
